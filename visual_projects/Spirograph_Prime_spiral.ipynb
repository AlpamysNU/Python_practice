## What is a spirograph?

This is a game/teaching tool ([Wikipedia link](https://en.wikipedia.org/wiki/Spirograph)).

To explain the trajectory of a point on a spinograph (often called a hypotrochoid, when a smaller circle rotates inside a larger circle), we will use the trigonometric description of the position of a point on the circle as a function of the angle of rotation. We have one larger circle with radius $ R $ and one smaller circle with radius $ r $ , and we want to trace the path of a point that is a distance $ d $ from the center of the smaller circle.

---

### Position of the center of the smaller circle

The center of the smaller circle rotates inside the larger circle. The position of the center of the smaller circle as a function of the angle $ \theta $ is given by the parametric equations in Cartesian coordinates:

- $ X = (R - r) \cos(\theta) $
- $ Y = (R - r) \sin(\theta) $

where $ \theta $ is the angle of rotation of the center of the smaller circle in radians.

### Rotation of the smaller circle

The smaller circle rotates (rolls around the larger circle). The angle by which the smaller circle has rotated about its own center relative to the starting point is given by the length of the arc it has covered:
$$ \theta R $$

if we want to express this arc relative to the angle of rotation in radians by which the smaller circle has rotated about its center, we convert back from circumference to angle:
$$ \theta \frac{ R}{r} $$
the last two important observations are:
- the small circle rotates negatively (so the angle is negative)
- the small circle rotates simultaneously with its center, so we must add the angle $\theta$ to the angle of rotation

Total we have:

$$ \phi = \theta - \theta \frac{R}{r} = - \theta \frac{R-r}{r} $$

### Position of the point on the spinograph

Finally, we combine the position of the center and the relative position of the point at a distance $d$ from the center of the smaller circle

- $ x = X + d \cos(\phi) = (R - r) \cos(\theta) + d \cos(- \theta \frac{R-r}{r})$
- $ y = X + d \sin(\phi) = (R - r) \sin(\theta) + d \sin(- \theta \frac{R-r}{r})$

If we use the properties (evenness/oddness) of trigonometric functions, we can simplify these equations to:

$$ x = (R - r) \cos(\theta) + d \cos(\theta \frac{R-r}{r})$$
$$ y = (R - r) \sin(\theta) - d \sin(\theta \frac{R-r}{r})$$

```python
#TODO
import numpy as np
import matplotlib.pyplot as plt

R = 7    
r = 2     
d = 5     

t = np.linspace(0, 2 * np.pi * r, 1000)


x = (R - r) * np.cos(t) + d * np.cos((R - r) / r * t)
y = (R - r) * np.sin(t) - d * np.sin((R - r) / r * t)


plt.figure(figsize=(6, 6))
plt.plot(x, y, linewidth=1)
plt.title("Trajectory")
plt.grid(True)
plt.show()
```

# 2. Prime spiral

**Part 1.: Drawing "sunflower points"**

- Draw the so-called "sunflower points". That is, the first `N` such points.
- Draw them so that the pattern is clearly visible (choose appropriate point sizes)

###### What are sunflower points?
The radius $ r $ and the angle $ \theta $ for each point (numbered by natural numbers: $n=1,...,N$) are determined as follows:

1. The golden angle $ \phi $ is given by the relation:
$$
\phi = \pi \cdot (3 - \sqrt{5})
$$

2. The radius $ r $ is derived from the index of point $ n $ as:
$$
r = \sqrt{n}
$$

3. The angle $ \theta $ for point $ n $ is calculated as:
$$
\theta = n \cdot \phi
$$

Each point is then located using Cartesian coordinates $(x, y)$, which are calculated using:
$$
x = r \cdot \cos(\theta)
$$
$$
y = r \cdot \sin(\theta)
$$

**Part 2: Plotting the Prime Spiral**
- Plot the prime spiral for the first `N` prime numbers.
- This is the same pattern as the "sunflower points" example, but this time we only plot the points that correspond to prime numbers.

```python
#TODO
import numpy as np
import matplotlib.pyplot as plt

N = 500

phi = np.pi * (3 - np.sqrt(5))

n = np.arange(1, N + 1)

r = np.sqrt(n)
theta = n * phi

x = r * np.cos(theta)
y = r * np.sin(theta)

plt.figure(figsize=(6, 6))
plt.scatter(x, y, s=5, color='orange')
plt.title("Sun flower Spiral)")
plt.axis('equal')
plt.grid(True)
plt.show()
```

```python
# Part 2
N = 5000

def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(np.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True

#    
primes = [i for i in range(1, N + 1) if is_prime(i)]
r_prime = np.sqrt(primes)
theta_prime = np.array(primes) * phi

x_prime = r_prime * np.cos(theta_prime)
y_prime = r_prime * np.sin(theta_prime)

plt.figure(figsize=(6, 6))
plt.scatter(x_prime, y_prime, s=8, color='blue')
plt.title("Prime Numbers Spiral)")
plt.axis('equal')
plt.grid(True)
plt.show()
```

# Bonus tasks
1. Modify the spinograph so that the smaller circle:
- rotates around the outside of the larger circle
- rotates around an equilateral n-gon (triangle, square, hexagon, ...)

```python
import numpy as np
import matplotlib.pyplot as plt

R = 5   
r = 3   
d = 1   

t = np.linspace(0, 2 * np.pi * r, 1000)

x = (R + r) * np.cos(t) - d * np.cos((R + r) / r * t)
y = (R + r) * np.sin(t) - d * np.sin((R + r) / r * t)

plt.figure(figsize=(6, 6))
plt.plot(x, y)
plt.title("Epitrochoid (Rolling Outside the Circle)")
plt.axis("equal")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def regular_ngon_vertices(n, radius):
    angles = np.linspace(0, 2*np.pi, n, endpoint=False)
    return np.column_stack((radius * np.cos(angles), radius * np.sin(angles)))

# - 
n = 6
R = 5
r = 1
d = 1
steps_per_edge = 100

#  
vertices = regular_ngon_vertices(n, R)
vertices = np.vstack((vertices, vertices[0]))  # 

#  
x_vals = []
y_vals = []

#   
for i in range(n):
    start = vertices[i]
    end = vertices[i+1]
    for j in range(steps_per_edge):
        t = j / steps_per_edge
        #    
        cx = (1 - t) * start[0] + t * end[0]
        cy = (1 - t) * start[1] + t * end[1]
        
        #  ""   
        angle = t * 2 * np.pi
        
        #   
        px = cx + d * np.cos(angle)
        py = cy + d * np.sin(angle)
        
        x_vals.append(px)
        y_vals.append(py)

# 
plt.figure(figsize=(6, 6))
plt.plot(x_vals, y_vals)
plt.title(f"Spirograph Rolling Along a Regular {n}-gon")
plt.axis("equal")
plt.grid(True)
plt.show()
```
